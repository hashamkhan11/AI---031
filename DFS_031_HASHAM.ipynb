**Objective: Implement search algorithms and apply them to solve problems.**

DFS Steps:

**Start at the first city:**

We begin from a specified city, such as Islamabad. We mark it as part of the current path.
**Explore one path as deep as possible:**

From Islamabad, we choose a neighboring city, like Rawalpindi, and continue exploring further from there.
If Rawalpindi has neighbors, we move to the next unvisited neighbor, such as Peshawar.
We keep moving to the next neighbor until we reach the destination city (Karachi), or there are no more neighbors to explore.

**Backtrack if necessary:**

If we reach a city (like Peshawar) that doesn’t lead to the destination or all its neighbors have been explored, we go back (backtrack) to the previous city.
We then try the next unvisited neighbor of the previous city.

**Continue until the destination is found:**

This process of moving forward (exploring) and backward (backtracking) continues until we find a path to the destination (Karachi).

**Store the path:**

Once we reach Karachi, we save this path. DFS doesn’t stop at finding one path—it keeps exploring to find all possible paths.

**Explore all paths:**

DFS will continue searching for alternative paths until all possibilities from the starting city (Islamabad) to the destination (Karachi) have been explored.

```python
import matplotlib.pyplot as plt  # Importing library for plotting

import networkx as nx  # Importing library for creating and manipulating graphs

def dfs(graph, start, goal, path=None):

    # If path is not provided, initialize it as an empty list

    if path is None:
        path = []

    # Add the starting node to the path
    path = path + [start]

    # If the start node is the goal, return the current path as a valid path
    if start == goal:
        return [path]

    # If the start node is not in the graph, return an empty list
    if start not in graph:
        return []

    # Initialize a list to hold all valid paths
    paths = []

    # Loop through each neighbor of the current node
    for neighbor in graph[start]:
        # If the neighbor has not been visited in the current path
        if neighbor not in path:
            # Recursively call dfs for the neighbor
            new_paths = dfs(graph, neighbor, goal, path)
            # Add any new paths found to the paths list
            for p in new_paths:
                paths.append(p)

    # Return all valid paths found
    return paths

# Define the graph representing connections between the characters
character_graph = {
    'Hasham': ['Subhan', 'Adan', 'Amina'],
    'Subhan': ['Hasham', 'Amina'],
    'Adan': ['Hasham', 'Amina'],
    'Amina': ['Hasham', 'Subhan', 'Areeba'],
    'Areeba': ['Amina']
}

# Find all paths from Hasham to Areeba using DFS
paths = dfs(character_graph, 'Hasham', 'Areeba')

# Print all paths found
print("All Paths from Hasham to Areeba:")
for path in paths:
    print(" -> ".join(path))  # Print each path in a readable format

def plot_graph(graph):
    # Create a new graph object
    G = nx.Graph()
    # Add edges for each character and their connections
    for character, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(character, neighbor)

    # Set up the figure for the plot
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)  # Define layout for the graph
    # Draw the graph with labels and styling
    nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=3000, font_size=10, font_weight='bold', edge_color='black')
    plt.title("Graph of Characters")  # Add a title to the plot
    plt.show()  # Display the plot

# Call the function to plot the graph of characters
plot_graph(character_graph)
```
