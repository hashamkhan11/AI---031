**Task 1: Basic Line Plot**

Create a basic line plot for the following data points
o x = [1, 2, 3, 4, 5]
o y = [2, 4, 6, 8, 10]

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')

plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)

plt.show()
```

**Task 2: Scatter Plot**

Create a scatter plot using the following data points:
o x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
o y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
Customize the markers with size and color.


```python
import matplotlib.pyplot as plt


x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, s=100, color='green', alpha=1, edgecolors='black')

plt.title('Customized Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)

plt.show()
```

**Task 3: Bar Plot**

Create a bar plot with the following categories and values:
o categories = ['A', 'B', 'C', 'D', 'E']
o values = [5, 7, 3, 8, 6]
Customize the bars with different colors for each category

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.bar(categories, values, color=colors)

plt.title('Customized Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

**Task 4: Histogram**

Generate a histogram for a dataset containing random numbers:
o Use np.random.randn(1000) to generate random numbers.
o Create a histogram with 30 bins and customize the color of the bars.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=1)

plt.title('Histogram of Random Numbers - 031')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

**Task 5: Pie Chart**

Create a pie chart for the following data:
o labels = ['Python', 'Java', 'C++', 'JavaScript']
o sizes = [50, 25, 15, 10]
Add percentages to each section and display the chart with an exploded view for the
largest section.


```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]

explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)

plt.title('Programming Language Popularity')

plt.axis('equal')

plt.show()
```
